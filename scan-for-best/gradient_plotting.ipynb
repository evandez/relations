```python
import matplotlib.pyplot as plt
import json
import numpy as np
import torch

plt.rcdefaults()
plt.rcParams["figure.dpi"] = 100
plt.rcParams["font.family"] = "Times New Roman"

SMALL_SIZE = 14
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc("font", size=SMALL_SIZE)  # controls default text sizes
plt.rc("axes", titlesize=BIGGER_SIZE)  # fontsize of the axes title
plt.rc("axes", labelsize=BIGGER_SIZE)  # fontsize of the x and y labels
plt.rc("xtick", labelsize=SMALL_SIZE)  # fontsize of the tick labels
plt.rc("ytick", labelsize=MEDIUM_SIZE)  # fontsize of the tick labels
plt.rc("legend", fontsize=SMALL_SIZE)  # legend fontsize
plt.rc("figure", titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
with open("corner_target_logit_sweep/sweep_loss_current.json") as f:
    gradient_performance = json.load(f)
```

```python
def get_performance_numbers(relation_id, gradient_performance = gradient_performance, weight = 'identity'):
    target_logit = []
    performance = []
    for t_logit in gradient_performance[relation_id]:
        if(t_logit == "jacobian_norm"):
            continue
        target_logit.append(float(t_logit))
        performance.append(gradient_performance[relation_id][t_logit][weight][0])
    return target_logit, performance
```

```python
weight = 'identity'
for relation_id in gradient_performance:
    target_logits, performance = get_performance_numbers(relation_id, weight=weight)
    plt.plot(target_logits, performance, label = relation_id)

plt.xlabel('target_logit')
plt.ylabel('precision @ 3')
plt.xticks(target_logits[::2])
plt.legend(bbox_to_anchor=(1.01, 1.01))
```

```python
def get_info_for_plotting(path, weight = 'identity'):
    with open(path) as f:
        gradient_performance = json.load(f)
    performance_track = []
    for relation_id in gradient_performance:
        target_logits, performance = get_performance_numbers(
            relation_id, gradient_performance= gradient_performance,
            weight=weight
        )
        performance_track.append(performance)
    return list(gradient_performance.keys()), np.array(performance_track)
```

```python
weight = "jacobian"
relation_ids, loss_current = get_info_for_plotting(
    "corner_target_logit_sweep/sweep_loss_current.json",
    weight=weight
)

relation_ids, loss_without_2nd_term = get_info_for_plotting(
    "corner_target_logit_sweep/sweep_loss_without_second_term.json",
    weight=weight
)

with open("corner_target_logit_sweep/new_loss.json") as f:
    cached_new_loss = json.load(f)
loss_new = np.array([cached_new_loss[r][weight][0] for r in relation_ids])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = (20,17)
plt.rcParams["figure.dpi"] = 200
plt.rcParams["font.family"] = "Times New Roman"

SMALL_SIZE = 14
MEDIUM_SIZE = 15
BIGGER_SIZE = 20

plt.rc("font", size=SMALL_SIZE)  # controls default text sizes
plt.rc("axes", titlesize=BIGGER_SIZE)  # fontsize of the axes title
plt.rc("axes", labelsize=BIGGER_SIZE)  # fontsize of the x and y labels
plt.rc("xtick", labelsize=SMALL_SIZE)  # fontsize of the tick labels
plt.rc("ytick", labelsize=BIGGER_SIZE)  # fontsize of the tick labels
plt.rc("legend", fontsize=BIGGER_SIZE)  # legend fontsize
plt.rc("figure", titlesize=BIGGER_SIZE)  # fontsize of the figure title


current_color = "teal"
without_second_term_color = "blue"
new_color = "darkorange"
bar_alpha = 0.6

y = np.arange(len(relation_ids))  # the label locations
width = 0.75  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.barh(y - width/3, loss_current.mean(axis = 1), width/3, label='loss_current', xerr = loss_current.std(axis = 1), color = current_color, alpha = bar_alpha)
rects2 = ax.barh(y, loss_without_2nd_term.mean(axis = 1), width/3, label=f'loss_without_second_term', xerr = loss_current.std(axis = 1), color = without_second_term_color, alpha = bar_alpha)
rects3 = ax.barh(y + width/3, loss_new, width/3, label='loss_new', color = new_color, alpha = bar_alpha)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_yticks(y, relation_ids)
ax.legend(loc="upper center", ncol=3)


def add_bar_labels(rects, accuracies, color = "black"):
    labels = [f"{round(acc.mean(), 2)} +/- {round(acc.std(), 2)}" for acc in accuracies]
    for rect, acc, label in zip(rects, accuracies, labels):
        x_pos = max(acc.mean() + acc.std(), acc.max()) + .01
        # x_pos = acc.max() + 0.01
        y_pos = rect.get_y() + rect.get_height() / 2
        ax.text(
            x_pos, y_pos, label, ha="left", va="center",
        )
        points = acc
        ax.scatter(points, [y_pos]*len(points), color = color, marker="|", s = 140)

add_bar_labels(rects1, loss_current, color = current_color)
add_bar_labels(rects2, loss_without_2nd_term, color = without_second_term_color)
ax.bar_label(rects3, padding=3)

fig.tight_layout()
plot_margin = 0.06
x0, x1, y0, y1 = plt.axis()
plt.axis((x0, x1 + plot_margin, y0, y1))
# plt.savefig(f"performance_conparison.png")
plt.gca().invert_yaxis()

fig.show()
```
